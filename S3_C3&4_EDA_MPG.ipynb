## Actividad 2. Dibujar un histograma y un boxplot de MPG

```python
#df['cylinders:'] = pd.to_numeric(df['cylinders:'], errors='coerce')

# Plot the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='cylinders:', y='mpg:')


# Title
plt.title('Boxplot of MPG by Cilinders')

# Show the plot
plt.show()
```

```python
# Plotting the histogram
df['mpg:'].hist(bins=20, figsize=(10, 6))

# Adding labels and title
plt.xlabel('MPG (Miles Per Gallon)')  # X-axis label
plt.ylabel('Frequency')  # Y-axis label
plt.title('Distribution of MPG')  # Title

# Display the plot
plt.show()
```
